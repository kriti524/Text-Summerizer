```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
from collections import Counter
from heapq import nlargest
import ipywidgets as widgets
from IPython.display import display

# Load the spaCy model
nlp = spacy.load('en_core_web_sm')

# Summarization function
def summarize_text(text, num_sentences=3):
    # Tokenization and removing stopwords
    doc = nlp(text)
    tokens = [token.text.lower() for token in doc
              if not token.is_stop and not token.is_punct and token.text != '\n']

    # Calculating word frequency
    word_freq = Counter(tokens)
    max_freq = max(word_freq.values())
    for word in word_freq.keys():
        word_freq[word] = word_freq[word] / max_freq

    # Sentence tokenization
    sent_token = [sent.text for sent in doc.sents]

    sent_score = {}
    for sent in sent_token:
        for word in sent.split():
            if word.lower() in word_freq.keys():
                if sent not in sent_score.keys():
                    sent_score[sent] = word_freq[word.lower()]
                else:
                    sent_score[sent] += word_freq[word.lower()]

    # Select top-scoring sentences
    summarized_sentences = nlargest(num_sentences, sent_score, key=sent_score.get)
    return " ".join(summarized_sentences)

# Setting up widgets for input and output
text_area = widgets.Textarea(
    value='Enter text here...',
    placeholder='Type something',
    description='Text:',
    layout=widgets.Layout(width="80%", height="200px")
)

num_sentences_slider = widgets.IntSlider(
    value=3,
    min=1,
    max=10,
    step=1,
    description='Sentences:',
    orientation='horizontal'
)

output_area = widgets.Output()

def on_summarize_button_click(b):
    with output_area:
        output_area.clear_output()
        summary = summarize_text(text_area.value, num_sentences=num_sentences_slider.value)
        print("Summary:")
        print(summary)

summarize_button = widgets.Button(description="Summarize")
summarize_button.on_click(on_summarize_button_click)

# Display the widgets
display(text_area, num_sentences_slider, summarize_button, output_area)
```

```python
import ipywidgets as widgets
from IPython.display import display
from transformers import pipeline

# Initialize the summarizer pipeline
summarizer = pipeline("summarization", model="t5-base", tokenizer="t5-base", framework="pt")

# Create an input text box widget
text_entry = widgets.Textarea(
    placeholder="Enter the text to summarize",
    description="Input Text:",
    layout=widgets.Layout(width="80%", height="200px")
)

# Create an output text box widget
output_text = widgets.Textarea(
    placeholder="Summary will appear here",
    description="Summary:",
    layout=widgets.Layout(width="80%", height="200px")
)

# Create a button to trigger text summarization
summarize_button = widgets.Button(description="Summarize")

# Function to summarize text and display output
def summarize_text(b):
    # Get text from the input text box
    text = text_entry.value

    # Summarize the text
    summary = summarizer(text, max_length=100, min_length=10, do_sample=False)

    # Update the output text box with the summary
    output_text.value = summary[0]['summary_text']

# Bind the button to the function
summarize_button.on_click(summarize_text)

# Display the widgets
display(text_entry, summarize_button, output_text)
```
